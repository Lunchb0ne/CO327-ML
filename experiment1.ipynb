Primitive types:

```
# int
a = 1
print(type(a))

# float
b = 2.22
print(type(b))

# bool
c = True
print(type(c))

# str
d = "wow"
print(type(d))
```

Container Types:

```
# list
l = ["x", 11 , 8.9]
print(type(l))

# tuple
t = (11,"y",7.4)
print(type(t))

# dict
di = {1:"one",3:"three",2:"two",3.14:"π"}
print(type(di))

# set
s = {1,9,3,0}
print(type(s))
```

Flow modifiers:

```
if(a>0):
    print("{} is positive".format(a))
else:
    print("{} is negative".format(a))
```

Slicing and other fun stuff:

```
lst=[10, 20, 30, 40, 50]

print("The whole list:", lst)

print("Length of the list:", len(lst))

print("The first element:", lst[0])

print("The list in reverse:", lst[::-1])

# we can access more subsequences by [start slice:end slice:step]
print("The list, but with steps of 2:", lst[::2])
```
